#### Uebung 7: Numpy & Matplotlib

```python
import matplotlib.pyplot as plt 
import numpy as np
```

##### Aufgabe 1

```python
x = np.random.uniform(-100, 100, 1000)
y = np.random.uniform(-100, 100, 1000)

farben = np.random.rand(1000,3)

for i in range(len(x)):
    plt.plot(x[i], y[i], 'o', color = farben[i])

plt.title("1000 Zufallszahlen")
plt.xlabel("x-Achse")
plt.ylabel("y-Achse")

plt.axis('equal')
plt.grid()
plt.show()
```

##### Aufgabe 2

```python
def f(x, y):
    return np.exp(-x**2) * np.sin(y)

x = np.linspace(-5, 5, 1000)
y = np.linspace(-5, 5, 1000)

X, Y = np.meshgrid(x, y)

Z = f(X, Y)

plt.pcolormesh(X, Y, Z)
plt.title("Auswertung der Funktion f(x,y)")
plt.xlabel("x-Achse")
plt.ylabel("y-Achse")

plt.axis('equal')
plt.grid()
plt.show()
```
